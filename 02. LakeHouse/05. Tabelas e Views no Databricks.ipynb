```sql
-- criando tabela CTA add configurações

CREATE or REPLACE TABLE logistica.cta_configuracoes_vendas_pneus
COMMENT 'CTA de teste venda pneus'
PARTITIONED BY (uf)
AS 
SELECT 
  id
  ,quantidade
  ,data_producao
  ,estado as uf 
FROM logistica.vendas_pneus 
WHERE estado = 'SP';
```
