```sql
%sql
-- Criando nova tabela com a versão 1
CREATE TABLE vendas_version_1 AS SELECT * FROM vendas VERSION AS OF 1;
```

```sql
%sql
-- Substituição da tabela atual por uma versão anterior
INSERT OVERWRITE TABLE vendas
SELECT * FROM vendas VERSION AS OF 1;
```

```sql
%sql
DESCRIBE HISTORY vendas;
```

```sql
%sql
-- criar view com uma versão expecifica
CREATE or REPLACE VIEW vw_vendas AS
SELECT * FROM vendas VERSION AS OF 2;
```

```sql
%sql
SELECT * FROM vw_vendas;
```
